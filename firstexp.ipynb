```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as splt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/new/heart.csv', sep=',')
```

```python
x = df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,13]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_curve, roc_auc_score
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=.3,random_state=1)
```

```python
import warnings
warnings.filterwarnings("ignore")
##  LogisticRegression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
lr_acc_score=accuracy_score(y_test, y_pred)
lr_acc_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
roc_score = roc_auc_score(y_test,y_pred)
roc_score
```

```python
tpr, fpr, thresholds = roc_curve(y_test, y_pred)
```

```python
splt.plot(tpr, fpr, color = 'blue', label = 'ROC')
splt.plot([0,1],[0,1],color = 'black', label = 'ROC curve (area = %0.2f)'% roc_score)
splt.xlabel("False Positivity Rate")
splt.ylabel("True Positivity Rate")
splt.title("Reciever Operator Characteristic curve")
splt.legend()
splt.show()
```

```python
##  KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
##  SVM
from sklearn.svm import SVC

model = SVC(kernel='linear')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
##  Naïve Bayes
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
##  Decision Tree
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
##  Random Forest
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
##  xgboost
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
splt.xlabel("Predicted")
splt.ylabel("Actual")
splt.title("Confusion Matrix")
splt.show()
```
